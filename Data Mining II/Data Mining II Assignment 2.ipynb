### Q1. Use the Decision Tree classification algorithm to construct a classifier on two datasets. Evaluate the classifier's performance by dividing the dataset into a training set (75%) and a test set (25%). Compare the performance with that of:
#### Bagging ensemble consisting of 3,5,7,9 Decision tree classifiers
#### Adaboost ensemble consisting of 3,5,7,9 Decision tree classifiers

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier,RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
d1=pd.read_csv("mushroom_cleaned.csv")
```

```python
y1=d1.iloc[:,8:9]
y1
```

```python
d2=pd.read_csv("drug200.csv")
d2
```

```python
from sklearn.preprocessing import LabelEncoder
def encode_labels(dataset):
    le = LabelEncoder()
    for column in dataset.columns:
        if dataset[column].dtype == 'object': 
            dataset[column] = le.fit_transform(dataset[column])
    return dataset
```

```python
d2_encoded = encode_labels(d2)
```

```python
d2_encoded
```

```python
x2=d2_encoded.iloc[:,:5]
x2
```

```python
y2=d2_encoded.Drug
y2
```

```python
def my_model(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

    dt = DecisionTreeClassifier(random_state=42)
    dt.fit(X_train, y_train)
    y_pred_dt = dt.predict(X_test)
    print("Decision Tree Classifier Accuracy:", accuracy_score(y_test, y_pred_dt))
    print(classification_report(y_test, y_pred_dt))

    # Bagging 
    for n_estimators in [3, 5, 7, 9]:
        bagging = BaggingClassifier(DecisionTreeClassifier(), n_estimators=n_estimators, random_state=42)
        bagging.fit(X_train, y_train)
        y_pred_bagging = bagging.predict(X_test)
        print(f"Bagging Classifier with {n_estimators} estimators Accuracy:", accuracy_score(y_test, y_pred_bagging))
        print(classification_report(y_test, y_pred_bagging))

    # AdaBoost
    for n_estimators in [3, 5, 7, 9]:
        adaboost = AdaBoostClassifier(DecisionTreeClassifier(), n_estimators=n_estimators, random_state=42)
        adaboost.fit(X_train, y_train)
        y_pred_adaboost = adaboost.predict(X_test)
        print(f"AdaBoost Classifier with {n_estimators} estimators Accuracy:", accuracy_score(y_test, y_pred_adaboost))
        print(classification_report(y_test, y_pred_adaboost))

# Evaluate for dataset1
print("Dataset 1 Results:")
my_model(x1, y1)

# Evaluate for dataset2
print("\nDataset 2 Results:")
my_model(x2, y2)
```

```python
datasets = {
    'Automobile': (X1, y1),
    'Heart Disease': (X2, y2),
    'Vehicle ': (X3, y3),
    'Glass': (X4, y4),
    'Sonar': (X5, y5),
    'Tic-Tac-Toe': (X6, y6),
    'Australian': (X7, y7),
    'Hepatatis': (X8, y8),
    'Iris': (X9, y9),
    'Wine': (X10, y10)}
```

```python
def Scoress(Model, test_labels, true_labels, Scores):
    accuracy = accuracy_score(test_labels, true_labels)
    precision = precision_score(test_labels, true_labels, average='weighted')
    recall = recall_score(test_labels, true_labels, average='weighted')
    f_measure = f1_score(test_labels, true_labels, average='weighted')
        
    Scores.append({
            'Model': Model,
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F-measure': f_measure
        })
```

```python
def model(datasetName, X, target):
    X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.25, random_state=42)

    Scores = []
    global Result

    # Decision Tree Classifier
    dcs = DecisionTreeClassifier(random_state=42)
    dcs.fit(X_train, y_train)
    label_1 = dcs.predict(X_test)
    Scoress('Decision Tree', y_test, label_1, Scores)

    # Bagging Classifier 
    bag = BaggingClassifier(estimator=dcs, n_estimators=5, random_state=42)
    bag.fit(X_train, y_train)
    label_2 = bag.predict(X_test)
    Scoress('Bagging Classifier', y_test, label_2, Scores)

    # AdaBoost Classifier
    adaboost = AdaBoostClassifier(estimator=dcs, n_estimators=5, random_state=42)
    adaboost.fit(X_train, y_train)
    label_3 = adaboost.predict(X_test)
    Scoress('AdaBoost Classifier', y_test, label_3, Scores)

    # Random Forest Classifier
    rf = RandomForestClassifier(random_state=42)
    rf.fit(X_train, y_train)
    label_4 = rf.predict(X_test)
    Scoress('Random Forest', y_test, label_4, Scores)

    # Store results in a DataFrame
    Metrics = pd.DataFrame(Scores)
    Metrics.insert(0, 'Dataset', datasetName)

    Result = pd.concat([Result, Metrics], ignore_index=True)

    return Result

# Initialize the result DataFrame
Result = pd.DataFrame(columns=['Dataset', 'Model', 'Accuracy', 'Precision', 'Recall', 'F-measure'])
```

```python
pd.set_option('display.max_rows', None)
for datasetName, (X, target) in datasets.items():
    model(datasetName, X, target)

Result
```
