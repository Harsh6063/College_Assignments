```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing First Datset
Survive=pd.read_csv("titanic.csv")
Survive.sample(5)
```

```python
Survive.info()
```

```python
#Removing The Unnessary Columns Not Required for Our Analysis
Survive_New=Survive.drop(['PassengerId','Name','SibSp','Parch','Embarked','Cabin','Ticket'], axis='columns')
Survive_New
```

```python
#Making Dummy Values(Removing categorical Values)
New_survive=pd.get_dummies(Survive_New.Sex)
New_survive=New_survive.astype(int)
```

```python
#Making The Final Dataset
Final_Survive=pd.concat([Survive_New,New_survive],axis='columns')
Final_Survive=Final_Survive.drop('Sex',axis="columns")
Final_Survive
```

```python
#Finding The Unnessary Values
Final_Survive.isnull().sum()
```

```python
# We See That Age Column has Unnessary values So Let's Fill Them With The Mean of Age Column
Avearge_Age=Final_Survive.Age.mean()
Final_Survive=Final_Survive.fillna(int(Avearge_Age))
Final_Survive
```

```python
#Checking Again
Final_Survive.isnull().sum()
```

```python
# Getting The Variables Assigned
inputs=Final_Survive[['Pclass','Age','Fare','female']]
output=Final_Survive.Survived
```

```python
#Importing Test_Train_Split For 75% Training sample
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,output,train_size=0.75,random_state=42) #For 75% Training sample
```

```python
#Importing Test_Train_Split For 66% Training sample
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test=train_test_split(inputs,output,train_size=0.66,random_state=42) #For 66% Training sample
```

```python
#Predicting Values
K_predict75.predict(x_test)
```

## Using random Subsampling

```python
# Making a Function to Splitting the dataset into Training and Testing Data

def split_data(test_size, random_state):
    from sklearn.model_selection import train_test_split
    X2_train, X2_test, y2_train, y2_test = train_test_split(inputs,output,test_size=test_size, random_state=random_state, shuffle=True)
    # Standardizing the Dataset
    from sklearn.preprocessing import StandardScaler
    sc = StandardScaler()# Splitting the dataset into Training and Testing Data
    X_train = sc.fit_transform(X2_train)
    X_test = sc.transform(X2_test)
    
    gnb = GaussianNB()
    y_pred = gnb.fit(X2_train, y2_train)

    y_pred_gnb = gnb.predict(X2_test)
    print("\nGaussian Naive Bayes model accuracy(in %):", accuracy_score(y2_test, y_pred_gnb)*100)
    accuracies_gnb.append(accuracy_score(y2_test, y_pred_gnb))
    
    # K-Nearest Neighbor Classifier

    knn = KNeighborsClassifier(n_neighbors=7)
    knn.fit(X2_train, y2_train)

    y_pred_knn = knn.predict(X2_test)
    print("K-Nearest Neighbors model accuracy(in %):", accuracy_score(y2_test, y_pred_knn)*100)
    accuracies_knn.append(accuracy_score(y2_test, y_pred_knn))
    
    # Decision Tree Classifier

    dtree = DecisionTreeClassifier()
    dtree.fit(X2_train, y2_train)

    y_pred_dtree = dtree.predict(X2_test)
    print("Decision Tree model accuracy(in %):", accuracy_score(y2_test, y_pred_dtree)*100)
    accuracies_dtree.append(accuracy_score(y2_test, y_pred_dtree))
    
    
```

```python
accuracies_gnb = []
accuracies_dtree = []
accuracies_knn = []
```

```python
from sklearn.metrics import accuracy_score
num = 42
for i in range(10):
    print(f"\nIter {i}:")
    split_data(test_size = 0.75, random_state = num)
    num += 10
    print("-"*100)   
```

```python
print("Final Accuracy of Naive Bayes of 75% Model: ", np.mean(accuracies_gnb)*100)
print("Final Accuracy of KNN of 75% Model: ", np.mean(accuracies_knn)*100)
print("Final Accuracy of Decision Tree of 75% Model: ", np.mean(accuracies_dtree)*100)
```

```python
from sklearn.metrics import accuracy_score
num = 42
for i in range(10):
    print(f"\nIter {i}:")
    split_data(test_size = 0.66, random_state = num)
    num += 10
    print("-"*100)   
```

```python
print("Final Accuracy of Naive Bayes of 66% Model: ", np.mean(accuracies_gnb)*100)
print("Final Accuracy of KNN of 66% Model: ", np.mean(accuracies_knn)*100) 
print("Final Accuracy of Decision Tree of 66% Model: ", np.mean(accuracies_dtree)*100)
```
